```python
from database import Database

class Users:
    def init(self, name, role):
        self.name = name
        self.role = role

class Orders:
    def init(self, user, tovar):
        self.user = user
        self.tovar = tovar

class Tovars:
    def init(self, name, price):
        self.name = name
        self.price = price

db = Database()

def register():
    name = input("Введите имя: ")
    role = input("Введите роль (client, employee, admin): ")
    if role not in ['client', 'employee', 'admin']:
        print("Некорректная роль")
        return
    db.add_user(Users(name, role))

def login():
    name = input("Введите имя: ")
    user = db.get_user(name)
    if user is None:
        print("Пользователь не найден")
        return
    password = input("Введите пароль: ")
    if password != user.password:
        print("Неверный пароль")
        return
    print("Добро пожаловать,", user.name)
    if user.role == 'client':
        client_menu(user)
    elif user.role == 'employee':
        employee_menu(user)
    elif user.role == 'admin':
        admin_menu(user)

def client_menu(user):
    while True:
        print("1. Просмотреть товары")
        print("2. Фильтрация товаров")
        print("3. Добавить товар в заказ")
        print("4. Удалить заказ")
        print("5. Изменить свои данные")
        print("6. Выйти")
        choice = input("Выберите действие: ")
        if choice == '1':
            tovars = db.get_tovars()
            for tovar in tovars:
                print(tovar.name, tovar.price)
        elif choice == '2':
            name = input("Введите название товара: ")
            tovars = db.filter_tovars(name)
            for tovar in tovars:
                print(tovar.name, tovar.price)
        elif choice == '3':
            name = input("Введите название товара: ")
            tovar = db.get_tovar(name)
            if tovar is None:
                print("Товар не найден")
                continue
            db.add_order(Orders(user, tovar))
            print("Товар добавлен в заказ")
        elif choice == '4':
            orders = db.get_orders(user)
            for i, order in enumerate(orders):
                print(i+1, order.tovar.name, order.tovar.price)
            choice = input("Выберите номер заказа для удаления: ")
            try:
                index = int(choice) - 1
                db.delete_order(orders[index])
                print("Заказ удален")
            except:
                print("Некорректный выбор")
        elif choice == '5':
            new_name = input("Введите новое имя: ")
            db.update_user(user, new_name)
            user.name = new_name
            print("Данные обновлены")
        elif choice == '6':
            break
        else:
            print("Некорректный выбор")

def employee_menu(user):
    while True:
        print("1. Просмотреть все товары")
        print("2. Фильтрация товаров")
        print("3. Добавить товар")
        print("4. Удалить товар")
        print("5. Изменить свои данные")
        print("6. Выйти")
        choice = input("Выберите действие: ")
        if choice == '1':
            tovars = db.get_tovars()
            for tovar in tovars:
                print(tovar.name, tovar.price)
        elif choice == '2':
            name = input("Введите название товара: ")
            tovars = db.filter_tovars(name)
            for tovar in tovars:
                print(tovar.name, tovar.price)
        elif choice == '3':
            name = input("Введите название товара: ")
            price = input("Введите цену: ")
            db.add_tovar(Tovars(name, price))
            print("Товар добавлен")
        elif choice == '4':
            name = input("Введите название товара: ")
            tovar = db.get_tovar(name)
            if tovar is None:
                print("Товар не найден")
                continue
            db.delete_tovar(tovar)
            print("Товар удален")
        elif choice == '5':
            new_name = input("Введите новое имя: ")
            db.update_user(user, new_name)
            user.name = new_name
            print("Данные обновлены")
        elif choice == '6':
            break
        else:
            print("Некорректный выбор")

def admin_menu(user):
    while True:
        print("1. Просмотреть всех сотрудников")
        print("2. Фильтрация сотрудников")
        print("3. Добавить сотрудника")
        print("4. Удалить сотрудника")
        print("5. Изменить свои данные")
        print("6. Выйти")
        choice = input("Выберите действие: ")
        if choice == '1':
            users = db.get_users()
            for user in users:
                print(user.name, user.role)
        elif choice == '2':
            name = input("Введите имя сотрудника: ")
            users = db.filter_users(name)
            for user in users:
                print(user.name, user.role)
        elif choice == '3':
            name = input("Введите имя сотрудника: ")
            role = input("Введите роль (employee, admin): ")
            if role not in ['employee', 'admin']:
                print("Некорректная роль")
                continue
            db.add_user(Users(name, role))
            print("Сотрудник добавлен")
        elif choice == '4':
            name = input("Введите имя сотрудника: ")
            user = db.get_user(name)
            if user is None:
                print("Сотрудник не найден")
                continue
            db.delete_user(user)
            print("Сотрудник удален")
        elif choice == '5':
            new_name = input("Введите новое имя: ")
            db.update_user(user, new_name)
            user.name = new_name
            print("Данные обновлены")
        elif choice == '6':
            break
        else:
            print("Некорректный выбор")

def main():
    while True:
        print("1. Регистрация")
        print("2. Авторизация")
        print("3. Выйти")
        choice = input("Выберите действие: ")
        if choice == '1':
            register()
        elif choice == '2':
            login()
        elif choice == '3':
            break
        else:
            print("Некорректный выбор")

if name == 'main':
    main()
```
